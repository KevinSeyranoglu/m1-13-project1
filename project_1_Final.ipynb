```python
import pandas as pd 
import numpy as np 
import matplotlib.pylab as plt 
from math import sqrt
from check_solution import check_solution
from check_solution import distance

#Notes on why I used the alpha beta gamma delta variables 
# I noticed that with those values I got better results 
#So I generate random values for those variables to see 
#if I can get a good result 

#the best solution I found gave me this result
# the bank list is in one of the cells 
#Time Remaining: 0.008840438025019791
#14770800.0

def bank_rob(dt,alpha,beta,gamma,delta):
    '''
    This function uses a greedy algorithm $/(travel time+rob time) with alpha,
    beta, gamma delta values that will help us find a good solution to The bank 
    robber algorithm. If we set alpha, beta, gamma, delta to 1 we have the original 
    greedy algorithm. 
    '''
    #this step is useless but I did not feel like 
    #changing 
    data=dt

    #travel speed
    v=30 #(km/h)

    #I'm going to start from the position (0,0) 
    #and visit the banks until I reach 24h. 

    #initializing the starting position 
    curent_x =0
    curent_y=0

    # x,y coordinate (array)
    x=data['x_coordinate']
    y=data['y_coordinate']

    #initializing time 
    t=0

    #initializing money
    money=0
        
    #list of the bank ids (initially empty )
    Bank_list=[]

    while t<=24:

        #distence with curent x,y position 
        d=np.sqrt((x-curent_x)**2+(y-curent_y)**2)
        
        #new travel_time comlumn 
        data['travel_time']=d/v

        # New Money_per_h column (greedy approach)
        data['Money_per_h']=data['money']/(alpha*data['time (hr)']**gamma+beta*data['travel_time']*delta)

        #finds the id of the maximum value of 'Money_per_h' 
        idmax=data['Money_per_h'].idxmax()

        #the row in witch we are interested
        interested_id=data.loc[idmax]

        #updated curent (x,y) position 
        curent_x=interested_id['x_coordinate']
        curent_y=interested_id['y_coordinate']

        #time it takes to go to the bank 
        t_travel=interested_id['travel_time']

        #time it takes to rob the bank 
        t_rob=interested_id['time (hr)']
        
        t+=(t_travel+t_rob)
    
        if t<24:
            money+=interested_id['money']

            #removing the bank from the data set 
            data=data.drop(index=interested_id['id'])

            #adding new bank id to the list
            Bank_list.append(int(interested_id['id']))

    #since we start at (x,y)=(0,0) we want to reverse the order 
    # or the robbed banks 
    Bank_list.reverse()

    data_test = pd.read_csv('bank_data.csv')

    return money, Bank_list

def bank_loop(dt,N):
    '''
    this function will run the bank_rob function with random values of alpha,
    beta, gamma, and delta for N try. The function will return the 
    highest amount of money obtained and the best values of alpha, beta,
    gamma, and delta we used to obtain the highest amount of money.

    This function uses a Monte-Carlo approche to find a good solution to 
    the The bank robber algorithm.
    '''

    #initializing empty lists 
    money_list=[]
    alpha_list=[]
    Beta_list=[]
    gamma_list=[]
    delta_list=[]

    for n in range(N):

        #ran num [0,1[
        alpha=np.random.random()
        beta=np.random.random()

        #--------------------------------------------
        #The intervals were chosen after many atempts
        #rand num [1,2[
        gamma=np.random.random()*2

        #rand num [0.8,1[
        delta=1-np.random.random()/10*5
        #--------------------------------------------

        m,banks=bank_rob(dt,alpha,beta,gamma,delta)

        #-------------------------
        #adding element to the lists 
        money_list.append(m)
        alpha_list.append(alpha)
        Beta_list.append(beta)
        gamma_list.append(gamma)
        delta_list.append(delta)
        #-------------------------
        print(m)
    #--------------------------------------------
    #list --> array
    money_list=np.array(money_list)
    alpha_list=np.array(alpha_list)
    Beta_list=np.array(Beta_list)
    delta_list=np.array(delta_list)
    gamma_list=np.array(gamma_list)
    #--------------------------------------------

    #array of the variables of the optimal value  
    al=alpha_list[np.where(max(money_list)==(money_list))]
    be=Beta_list[np.where(max(money_list)==(money_list))]
    ga=gamma_list[np.where(max(money_list)==(money_list))]
    de=delta_list[np.where(max(money_list)==(money_list))]

    return max(money_list),al,be,ga,de
```

```python
import time

t0 = time.time()
#number of time we run the algorithm 
N=1000
dt = pd.read_csv('bank_data.csv')
money,a,b,g,d=bank_loop(dt,N)

print('max money=',money)
print('Alpha=',a),
print('Beta=',b)
print('Delta=',d)
print('gamma=',g)
t1 = time.time()
total = t1-t0
print(total )



```
